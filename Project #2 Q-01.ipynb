## Happiness 2015

```r
happy15 <- read.csv("https://raw.githubusercontent.com/sauchilee/Stat517/master/Data/World_Happiness_2015.csv")
```

```r
head(happy15, n=90)
```

```r
names(happy15) # Displays the attributes
```

```r
hp15 <- happy15[6:12] # Removing happiness information
hp15
```

### K-means clustering

```r
set.seed(123)
# Compute and plot wss for k = 2 to k = 15
k.max <- 15 # Maximal number of clusters
data <- hp15
wss <- sapply(1:k.max, 
        function(k){kmeans(data, k, nstart=10 )$tot.withinss})

plot(1:k.max, wss,
       type="b", pch = 19, frame = FALSE, 
       xlab="Number of clusters K",
       ylab="Total within-clusters sum of squares")
abline(v = 3, lty =2)
```

From this graph, we can see that 3 clusters are optimal for k-means clustering

```r
results15 <- kmeans(hp15, 3)
results15
```

```r
table(hp15$Economy..GDP.per.Capita., results15$cluster)
# Displaying the clusters for each record, grouped by the GDP per capita for that country
```

## Displaying the clusters using plots

```r
plot(hp15[c("Economy..GDP.per.Capita.","Health..Life.Expectancy.")], col = results15$cluster)
```

```r
plot(hp15[c("Economy..GDP.per.Capita.","Health..Life.Expectancy.")], col = happy15$Happiness.Score)
# Comparing the previous plot to the actual, to evaluate the difference while clustering
```

From these plots, it is evident that the generated plot has not deviated much from the original. This indicates that the clustering analysis was adequate.

## Gaussian Mixture: Happiness 2015

```r
hp15_clust <- Mclust(hp15)
names(hp15_clust)
```

```r
hp15_clust$classification
```

The countries were classified into four clusters generated using gaussian mixture. The corresponding countries were identified from the original dataset. Few countries from each cluster are displayed below.

```r
country_clust1 <- data.frame(Cluster1 = c("Switzerland", "Iceland", "Denmark", "Norway", "Canada", "Finland", "Netherlands"), 
                 Cluster2 = c("United States", "Luxembourg", "United Arab Emirates", "Oman", "Singapore", "Qatar", "Uruguay"), 
                 Cluster3 = c("Israel", "Costa Rica", "Mexico", "Brazil", "Venezuela", "Panama", "Chile"), 
                 Cluster4 = c("Suriname", "Bolivia", "Togo", "Burundi", "Benin", "Afghanistan", "Burkina Faso"))
country_clust1
```

From these results we can see that the method was effective in clustering the countries based on raw data, without the happiness information. Almost all countries in cluster1 has the highest happiness ranking, whereas the countries in cluster 4 mostly has the lowest happiness ranking.

```r
summary(hp15_clust)
```

```r
plot(hp15_clust, what = 'classification', dimens = c(1, 2))
```

```r
hp15_clust$z
```

The above table displays the probability of each datapoint being sorted into each cluster

```r
plot(hp15_clust) # Identifying the best model for this dataset
```

From these plots, we can see that VVE with 4 components is the best model.

```r
hp15_BIC <- mclustBIC(hp15)
hp15_BIC
```

## Seriation: Happiness 2015

```r
library(seriation)
a <- as.matrix(hp15)
d <- dist(a)
s <- seriate(d)
pimage(d, main='original')
pimage(d, s, main='reordered')
```

```r
get_order(s) # Generating the new order based on the seriation analysis
```

According to the order generated by seriation for the happiness 2015 dataset, Mexico (14) was found to be ranked number 1, followed by Central African Republic (148). Norway (4) was placed in the 29th position in the generated ranklist. The top ten countries from the raw data, does not follow the same order after seriation analysis. China seems to have fallen back in the order that was generated. From the order, we can see that China is no happier than most countries, even though they are rich and have longer longetivity. It is clear that this generated ranklist is different from the original ranklist. 

```r
hmap(a)
```
